```python
from datasets import load_dataset
dataset = load_dataset("microsoft/ms_marco", "v2.1", split="train", streaming=True)
```

```python
documents = []
max_doc = 5

for example in dataset:
    list_passages = example["passages"]["passage_text"]
    for passage in list_passages:
        if len(documents)<max_doc:
            proper_text = passage.replace("\n", " ").strip()
            documents.append(proper_text)
        else:
            break
    if len(documents)>=max_doc:
        break
    
            

```
